```python
# Problem 1: Social Media – Post Engagement Analyzer

def post_engagement_analyzer(likes):
    total_likes = 0
#loop to calculate total likes
    for like in likes:
        total_likes += like
#check post status
    if total_likes >=1000 :
        status = "Viral Post"
    else:
        status = "Normal Engagemnet"
    
    print("total likes:", total_likes) 
    print("Status:", status)
#input
likes = [50,150,700,110,40]
post_engagement_analyzer(likes)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker

def analyze_rainfall(rainfall_data, required_level):
    total = 0 #  creating varaible to store total value
    
    for rain in rainfall_data:
        total += rain
    #calcuating Average
    average = total / len(rainfall_data)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

# Example usage
data = [70, 75, 68, 80, 67]
required = 70

analyze_rainfall(data, required)
```

```python
# Problem 4: Social Media – Duplicate Account Detection

def detect_duplicate_users(usernames):
    #convert list to set to remove duplicates
    unique_users =set(usernames)

#compare lengths to check duplicates
    if len(unique_users)!= len(usernames):
        result = "Yes" 
    else:
        result = "No"
    print("Duplicate Accounts Found:", result)
 #Example list of usernames
users = ["ravi123", "anita_01", "ravi123", "kiran99"]
#call the fucntion
detect_duplicate_users(users)
    
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices):
    premium = []
    #Loop through each prize
    for price in prices:
        if price > 2000:
            premium.append(price)
    print("Premium Crops:",premium)
#exmple
crop_prices = [1800,2500,1990,3200,1500]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker

def check_system_health(error_count):
# Determine system status based on error count
    if error_count == 0:
        status ="Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("Error Count:", error_count)
    print("System Status:", status)
#Example usage
check_system_health(7)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker

def validate_transaction(amount):
    Daily_limit = 50000
    #check if amount is within the allowed limit
    if amount <= Daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)
#example usage
validate_transaction(60000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0 #counter for present days
#loop through attendance list
    for status in attendance_list:
        if status == "P":
            present_days += 1
            # calcualte attendance percentage
    percentage = (present_days / total_days )* 100
#checking eligibility'
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

#input/ example usage
attendance = ["P", "A", "P", "P", "P"]
check_exam_eligibility(attendance)
    
    
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
# creating variable To store total bill
    bill = 0 
#calculate bill based on slab system
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill =(100 * 3) + (100 * 5) + (units - 200) *  7
#usage amount Based on bill amount
    if bill < 500:
        usage = "Moderate Usage"
    elif bill <= 1500:
        usage = "Moderate usage"
    else:
        usage = "High Usage"
    return bill, usage
#Example usage
units_consumed = 400 
bill_amount, usage_status = calculate_electricity_bill(units_consumed)

print("Units consumed:", units_consumed)
print("Total Bill:", bill_amount)
print("Usage level:", usage_status)
    
```
