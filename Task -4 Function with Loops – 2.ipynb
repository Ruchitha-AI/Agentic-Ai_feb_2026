```python
##### Problem 1: Mobile Recharge Validation System

def validate_recharge_request(amount):
    valid_plans = [199, 299, 399, 599]
    if amount <= 50:
       print("minimun recharge amount is 50")
       return False
    elif amount not in valid_plans:
        
         print("Invalid plan selected.")
         print("Available plans:", valid_plans)
         return False
    else:
         print("Recharge successful!")
         return True
    
#using while loop for retry
while True:
    amount = int(input("Enter recharge amount:"))
    if validate_recharge_request(amount):
                 break
    else:
        print("please try again")
                 
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_cab_fare(distance,peak):
    #base fare is 50
    fare = 50 + (12* distance)
    #checks if ride durin peak hours
    if peak == "yes":
        fare += fare*0.25     #25% extra for peak hour
    return fare
#loop for calculatong fare multiple times
while True:
    #taking input from user
    distance = float(input("Enter distance (km):"))
    peak = input("peak hour? (yes/No):")
    
    print("Fare = ₹", calculate_cab_fare(distance,peak))
# asking user if they want to calculate again
    retry = input("Calculate again? (yes/No): ")
    #if user give input other than  yes,stop the loop
    if retry != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def  check_attendance(att_list):
    present_days = 0
#count present days
    for status in att_list:
        if status =="P":
            present_days += 1
    total_days = len(att_list)
    if total_days == 0:
        return "No attendance data"
#calculate attendance percentage
    percentage = (present_days / total_days) * 100
#check eligibility
    if percentage >= 75:
        return"Eligible"
    else:
        return"Not Eligible"
#example
attendance = ["P","A","P","P","P","A","P","P"]
result = check_attendance(attendance)

print("Attendance percentage:",(attendance.count("P") / len(attendance)) * 100)
print("Status:", result)
        
```

```python
# Problem 6: Password Strength Checker

def check_password_strength(password):
    has_digit = False
    has_specil = False
    #loop through each character in password
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in  "@$":
            has_special = True
    #check all rules 
    if len(password) >= 8 and has_digit and  has_special:
        return"Strong Password"
    else:
        return"Weak Password"
#user input
password = input("Enter password:" )
#function call
result = check_password_strength(password)
print(result)
```
